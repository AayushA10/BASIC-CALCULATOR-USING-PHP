```python

import pandas as pd
import matplotlib.pyplot as plt


data_file_path = "data.csv"
air_quality_data = pd.read_csv(data_file_path, encoding='latin1')


air_quality_data['date'] = pd.to_datetime(air_quality_data['date'], errors='coerce')
air_quality_data['year'] = air_quality_data['date'].dt.year
air_quality_data['month'] = air_quality_data['date'].dt.month


monthly_trends = air_quality_data.groupby('month')[['so2', 'no2', 'rspm', 'spm', 'pm2_5']].mean()


plt.figure(figsize=(12, 8))
monthly_trends.plot(marker='o', figsize=(12, 6))
plt.title("Seasonal Air Quality Trends (Average Monthly Levels)", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Pollutant Levels (Mean)", fontsize=12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title="Pollutants", fontsize=10)
plt.grid()
plt.show()
```

```python

state_pollutant_levels = air_quality_data.groupby('state')[['so2', 'no2', 'rspm', 'spm', 'pm2_5']].mean().sort_values(by='so2', ascending=False)


plt.figure(figsize=(14, 8))
state_pollutant_levels.plot(kind='bar', figsize=(14, 8))
plt.title("Pollutant Levels Across Indian States", fontsize=16)
plt.xlabel("State", fontsize=12)
plt.ylabel("Mean Pollutant Levels", fontsize=12)
plt.xticks(rotation=90)
plt.legend(title="Pollutants", fontsize=10)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
